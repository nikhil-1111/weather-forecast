```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report
import pickle
```

```python
df = pd.read_csv(r"weather_forecast_data.csv")
df
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
x = df.iloc[:,[0,1,2,3,4]].values
print(pd.DataFrame(x))
```

```python
y = df.iloc[:,-1].values
print(pd.DataFrame(y))
```

```python
lab = LabelEncoder()
y = lab.fit_transform(y)
print(pd.DataFrame(y))
```

```python
x_train, x_test, y_train , y_test = train_test_split(x, y, test_size=0.25, random_state= 42)
```

```python
print(x_train.shape, y_train.shape)
```

```python
print(x_test.shape, y_test.shape)
```

```python
rf = RandomForestClassifier(random_state=42, n_estimators = 100, max_depth = 10)
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
print(pd.DataFrame({"actual":y_test,"Predicted":y_pred}).to_string())
```

```python
print("accuracy",accuracy_score(y_test,y_pred)*100)
```

```python
class_label = {1:'rain' , 0:'no rain'}
print(classification_report(y_test,y_pred))
```
